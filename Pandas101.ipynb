HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ผู้ชาย ของข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ผู้ชาย ของผู้ป่วยในจังหวัดขอนแก่น

````
this_data = data_covid[['sex','age','province_of_onset']]  #เลือกมาเฉพาะ column ที่ต้องการ
````

````
this_data
````

````
female = this_data[this_data['sex']=='หญิง'] #สนใจข้อมูลที่เป็นเพศหญิง
female
````

````
sum = 0 #ให้เป็น 0 เพราะเริ่มต้นถ้าไม่มีอะไรบวกเข้ามามันต้องเป็น 0 ก่อน ตัวที่ 1บวกเข้ามาก็เป็นค่าตัวที่ 1 พอตัวที่ 2 เข้ามาก็เป็นค่าตัวที่ 2+1 ฉะนั้นมันต้องเริ่มจากค่าที่ 0
#ตัวแปร sum = 0 คือตัวแปร sum จะมารับค่าทั้งหมด ที่จะเอามาบวกกัน เป็นคุณสมบัติการบวก
#ถ้าเอาตัวแปรมารับค่าที่คูณกันต้องเริ่มจาก 1 ถ้าเริ่ม 0 มันก็ 0ไปหมด 
N = 0 # N นับว่าเป็นกี่คน 
for a in female['age']:
    
    if a > 0:  #เขียนวนลูป เชคก่อนถ้ามากกว่า 0 ก็เป็นอายุของคนปกติ ที่มากกว่า 0
        sum += a # sum = sum + a เอา a มาบวกลงไปใน sum เรื่อยๆถ้า a>0
        N += 1 #คำนวณจำนวณด้วย 
 #พอวนลูปครบก็จะได้ sum = ผลรวมของอายุทั้งหมด N=จำนวนคนทั้งหมดที่บวกรวมในsum
print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง {sum/N}')
````

````
female_KK = female[female['province_of_onset']=='ขอนแก่น']  #สนใจเฉพาะข้อมูลที่เป็นเฉพาะจังหวัดขอนแก่น

sum = 0
N = 0
for a in female_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง ในจังหวัดขอนแก่น {sum/N}')
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

###HW กลุ่ม 2  เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

````
AN = list()  #ประกาศ AN เป็น list ว่าง
for each_row in data_covid.iterrows():  #เขียนลูปในตาราง data_covid 
    if (int(each_row[1]['announce_date'].split('/')[1] ) >=1) and (int(each_row[1]['announce_date'].split('/')[1] ) <=20):  
        AN.append(True)  #ตรงเงื่อนไข  = true
    else:
        AN.append(False)   #ไม่ตรงเงื่อนไข  = false

data_covid[AN]
````
