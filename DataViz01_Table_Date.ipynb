```python
import pandas as pd #import panda เพื่อจัดการตารางเล็กน้อย
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

```python
df['Name'].unique() #unique=เลือกมาเฉพาะคอลัมน์ที่ไม่ซ้ำ
```

image.png

###Parallel Coordinates

คือเราจะเอาค่าของแต่ละ column มา plot เทียบกัน

```python
pd.plotting.parallel_coordinates(df, 'Name') #Parallel Coordinates คือเราจะเอาค่าของแต่ละ column มา plot เทียบกัน โดยใช้ column name เป็นตัวแบ่ง
#เส้นนึง = Data 1 จุด
#Virginica ใหญ่กว่า verciclor แบบ กลีบดอกใหญ่ไรงี้
```

```python
pd.plotting.parallel_coordinates(df, 'Name', cols = ['SepalLength','PetalLength']);
```

###change color

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['r','y','k'])
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])
```

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['#FF00E0','#00FFE2','ECFF00'])
```

###Special color by colormap
กลุ่มสีที่เค้ากำหนดไว้อยู่แล้ว

```python
pd.plotting.parallel_coordinates(df, 'Name', colormap = 'Set1')
```

###Subplot

```python
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
plt.subplot(1,3,1) #subplot(มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot')
```

```python
plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha=0.2) #alpha กำหนดความโปร่งใสของรูป alpha=1=ไม่โปร่งใสเลย

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha=0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha=0.8)
```
