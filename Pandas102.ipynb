####Dictionary (row oriented)

```python
#สร้าง list ขึ้นมา 1 ตัว  เพื่อเอา list ยัดเข้าไปในตัวนั้น  key=column value=ค่าข้างใน
data_list = [ {'account':'Jones LLC','Jan':150,'Feb':200,'Mar':140},   #row แรก
              {'account':'Alpha Co','Jan':200,'Feb':210,'Mar':215},  #row ที่ 2 
              {'account':'Blue Inc','Jan':50,'Feb':90,'Mar':95},
             ] 
```

```python
data_list
```

```python
df1 = pd.DataFrame(data_list)
df1
```

####List (row oriented)

```python
data = [['Jones LLC',150,200,140],  #list ที่เป็น Data เราเอาค่าในแต่ละแถวมา 
        ['Alpha Co',200,210,215],
        ['Blue Inc',50,90,95]]
header = ['account','Jan','Feb','Mar']  #หัวตาราง 
```

```python
df2 = pd.DataFrame.from_records(data,columns=header)
df2
```

####Dictionary (column oriented)

```python
data_dict = { 'account':['Jones LLC','Alpha Co','Blue Inc'],  #เราสร้าง list ตามแนว row/column ก็ได้
              'Jan':[150,200,50],
              'Feb':[200,210,90],
              'Mar':[140,215,95]
}
```

```python
df3 = pd.DataFrame.from_dict(data_dict)
df3
```
